## Characteristic strain for LISA

```python
# Smith -- LISA for cosmologists paper

f = np.linspace(10e-5,1) #Hz
# f_mHz = f * 10e-3
f_1 = 4e-7 #.4 mHz
f_2 = 2.54e-6 #2.54 mHz
R_f = (f/f_2)**2
S_II_f = 3.6e10-41 # 1/Hz
S_I_f = 5.76 * 10e-48 * (1 + (f_1/f)**2) # 1/s^4 * 1/Hz
```

```python
Sigma_h = (20/6) * ( (S_I_f / (2*math.pi*f)**4) + S_II_f ) * R_f
sensitivity = 2 * np.sqrt(f * Sigma_h)
```

```python
fig, ax = plt.subplots()
ax.set(xlim=(10e-4,10e2))
ax.set_yscale('log')
plt.plot(Sigma_h)
ax.set_xscale('log')
```
